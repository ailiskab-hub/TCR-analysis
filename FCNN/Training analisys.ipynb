```python
import pandas as pd
```

```python
losses = pd.read_csv('losses.csv')
losses
```

```python
scores = pd.read_csv('scores.csv')
scores
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# losses plot
epochs = range(1, 30 + 1)
plt.figure(figsize=(12, 8))

for k, v in losses.items():
    plt.plot(epochs, v, '.-', label=k)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
    
```

```python
scores.loc[0]
```

```python
n_iters = 72
v_score = scores.loc[0]
j_score = scores.loc[1]
    
barWidth = 0.4
fig = plt.subplots(figsize =(18, 10)) 
    
br1 = np.arange(n_iters) 
br2 = [x + barWidth for x in br1] 

plt.bar(br1, v_score, color ='#9467bd', width = barWidth, label ='V accuracy') 
plt.bar(br2, j_score, color ='#4d9221', width = barWidth, label ='J_accuracy') 

plt.xlabel('Models') 
plt.ylabel('Accuracy') 
plt.xticks([r + barWidth/2 for r in range(n_iters)], scores.keys(), rotation=90)

plt.legend()
plt.savefig('../imgs/FCNN/score_all_beta.png', bbox_inches='tight')

plt.show()
```

```python
losses_df_best = losses.T[losses.T[29]<4.3]
```

```python
losses_df_best
```

```python
# losses plot
epochs = range(1, 30 + 1)
plt.figure(figsize=(12, 8))

for i in range(len(losses_df_best.values)):
    plt.plot(epochs, losses_df_best.values[i], '.-', label=losses_df_best.index[i])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.savefig('../imgs/FCNN/best_losses_beta.png', bbox_inches='tight')
plt.show()
```

```python
scores_df_best = scores.T[scores.T[0]>0.75]
```

```python
scores_df_best.sort_values(by=0, ascending=False)
```

```python
n_iters = scores_df_best.shape[0]
v_score = scores_df_best[0]
j_score = scores_df_best[1]
    
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 
    
br1 = np.arange(n_iters) 
br2 = [x + barWidth for x in br1] 

plt.bar(br1, v_score, color ='#9467bd', width = barWidth, label ='V accuracy') 
plt.bar(br2, j_score, color ='#4d9221', width = barWidth, label ='J_accuracy') 

plt.xlabel('Models') 
plt.ylabel('Accuracy') 
plt.xticks([r + barWidth/2 for r in range(n_iters)], scores_df_best.index, rotation=90)

plt.legend()
plt.savefig('../imgs/FCNN/best_scores_beta.png', bbox_inches='tight')

plt.show()
```
